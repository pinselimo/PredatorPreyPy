# Notebook for analysis of the PredatorPreyPy package

```python
import numpy as np
import predatorpreypy as ppp
import matplotlib.pyplot as plt
```

```python
def plot_run(ticks=10):
    env = ppp.Environment((10,10), 50, 50)
    results = list()
    for _ in range(ticks):
        env.tick()
        results.append(
            (len(env.predators), len(env.prey))
        )
    plt.plot(results)
```

```python
plot_run()
```
